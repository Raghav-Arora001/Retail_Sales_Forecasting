## Business Problem

Rossmann operates thousands of drug stores across Europe. Accurate sales forecasting helps the company optimize inventory, staff scheduling, promotional planning, and revenue forecasting.

The objective of this project is to build a machine learning model capable of predicting daily sales for each Rossmann store using historical sales data and store-related information.

## Dataset Files

This project primarily uses two datasets:

- **train.csv** – Historical daily sales data containing the target variable (`Sales`).
- **store.csv** – Store-level information such as store type, assortment, competition details, and promotion participation.

Although the original Kaggle dataset also includes `test.csv` and `sample_submission.csv`, they are intended for the competition workflow. Since this project focuses on developing and evaluating machine learning models using classical ML techniques, model validation will be performed by splitting `train.csv` into training and validation sets.
