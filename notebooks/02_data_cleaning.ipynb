# Phase 2: Data Cleaning

## Objective

The goal of this notebook is to prepare a clean, consistent dataset for exploratory data analysis (EDA) and feature engineering.

In this phase, we will:

- Merge the sales and store datasets.
- Validate the merge operation.
- Convert the `Date` column to the appropriate datetime format.
- Audit missing values and classify their causes.
- Create a dataset containing only open stores for model training.
- Save the cleaned datasets for use in later phases.

### Important Principles

- Raw datasets remain unchanged.
- Every preprocessing decision must be justified.
- Missing values will be understood before being treated.
- Cleaning is separated from feature engineering.

```python
import pandas as pd
import numpy as np
```

```python
train_df=pd.read_csv("../data/raw/train.csv")
store_df=pd.read_csv("../data/raw/store.csv")
```

```python
merged_df=train_df.merge(store_df,on="Store",how="left")
```

```python
missing_values = merged_df.isnull().sum()
missing_values = missing_values[missing_values > 0]
missing_values.sort_values(ascending=False)
```

## Merge Validation

The sales dataset was merged with the store metadata using a left join on the `Store` column.

Validation checks confirmed that:

- The number of rows remained unchanged.
- No duplicate records were introduced.
- Every sales record successfully matched with store metadata.
- The merged dataset contains 18 columns.

These checks confirm that the merge was successful and no information was lost.

## Creating the Open Store Dataset

The project aims to forecast sales for stores that are operating.

Closed stores always record zero sales because no transactions can occur. Including these observations in the training dataset would encourage the model to learn the trivial relationship between the `Open` indicator and sales, rather than learning the effects of promotions, seasonality, holidays, and competition.

To focus the model on demand forecasting, a separate dataset containing only open-store observations is created.

```python
train_open_df = merged_df[merged_df["Open"]==1].copy()
```

```python
merged_df.to_csv("../data/processed/merged_data.csv", index=False)
train_open_df.to_csv("../data/processed/train_open.csv", index=False)
```

## Saving Processed Datasets

Two processed datasets were saved:

- `merged_data.csv` — complete cleaned dataset for exploratory analysis and reporting.
- `train_open.csv` — dataset containing only open-store observations for feature engineering and model development.

Saving processed data avoids repeating preprocessing in future notebooks and improves the reproducibility of the workflow.
